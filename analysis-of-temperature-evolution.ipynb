## Filtragem Mediana

A ```filtragem mediana``` é uma técnica de filtragem digital não linear, frequentemente usada para remover ruídos de uma imagem ou sinal. A ideia principal do filtro de medianas é percorrer o sinal entrada por entrada, substituindo cada entrada pela mediana das entradas vizinhas. A mediana é menos sensível a pixels de valores extremos e o filtro de mediana é bastante eficiente em imagens de radar para suprimir um tipo muito comum de ruído que estas imagens apresentam, chamado de speckle (ruído sal e pimenta).

Podemos utilizar a biblioteca ```scipy``` e o subpacote ```ndimage``` e o método **filtro médiano** para reduzir o ruído no sinal.


````python
from scipy.ndimage import median_filter

data = sinal

filtered_data = median_filter(data, size=6)

print(filtered_data)

plt.figure(figsize=(10, 5))

plt.plot(filtered_data)

plt.xlabel('Sinais')

plt.show()
````

## Filtragem Gaussiana

A ```filtragem gaussiana``` é uma técnica de suavização de dados que pode ser aplicada a listas, assim como a imagens e outros tipos de dados. A ideia é substituir cada valor da lista por uma média ponderada dos valores vizinhos, onde os pesos são determinados por uma ```função gaussiana```. Isso tem o efeito de suavizar os dados, reduzindo o ruído e preservando as características gerais da lista.

Você pode implementar a filtragem gaussiana em listas usando bibliotecas como ```NumPy``` e ```SciPy```.

````python
import numpy as np

input_list = sinal

sigma = 5

kernel_size = int(2 * np.ceil(2 * sigma) + 1)
kernel = np.exp(-np.arange(-kernel_size // 2, kernel_size // 2 + 1) ** 2 / (2 * sigma ** 2))
kernel /= kernel.sum()

output_list = np.convolve(input_list, kernel, mode="same")

print(output_list)
````

Agora vamos exibir esses dados em um gráfico com ```matplotlib```.

````python
plt.figure(figsize=(10, 5))

plt.plot(output_list)

plt.xlabel('Sinais')

plt.show()
````

## Filtragem de Butterworth

O ```filtro Butterworth``` é um tipo de projeto de filtros eletrônicos. Ele é desenvolvido de modo a ter uma resposta em frequência o mais plana o quanto for matematicamente possível na banda passante. Os ```filtros Butterworth``` foram descritos primeiramente pelo engenheiro britânico ```Stephen Butterworth``` na sua publicação “On the Theory of Filter Amplifiers”. Embora o **filtro Butterworth** seja mais comumente usado em aplicações de processamento de sinal, ele também pode ser aplicado a listas de dados para suavizar os valores e reduzir o ruído.

````python
from scipy.signal import butter, lfilter

input_list_butter = sinal

order = 3

cutoff = 0.1

b, a = butter(order, cutoff)

output_list_butter = lfilter(b, a, input_list_butter)

print(output_list_butter)
````

````python
plt.figure(figsize=(10, 5))

plt.plot(output_list_butter)

plt.xlabel('Sinais')

plt.show()
````

# Transformada de Fourier

Outra técnica é usar a transformada de Fourier.

A **Transformada de Fourier** é uma ferramenta matemática que permite decompor um sinal em seus componentes de frequência. Ela converte um sinal do domínio do tempo para o domínio da frequência, fornecendo uma representação do sinal em termos de senos e cossenos de diferentes frequências. A ```Transformada de Fourier``` é amplamente utilizada em áreas como **processamento de sinais**, **análise espectral**, **compressão de dados** e **resolução de equações diferenciais**. Ela permite analisar as propriedades espectrais de um sinal, como sua largura de banda e conteúdo harmônico, e também pode ser usada para filtrar sinais e remover ruído. Existem várias versões da transformada de Fourier, incluindo a **transformada contínua de Fourier** (CFT), a **transformada discreta de Fourier** (DFT) e a **transformada rápida de Fourier** (FFT).
